```python
import os
from panopy import read_navvis_alignment_xml, read_navvis_pano_poses_csv, plot_pose_collections
from pathlib import Path
```

```python
dir=Path(os.getcwd()).parents[1].__str__()+'\Samples\PanoPositioning/'
filename = dir +'alignment.xml'
```

```python
# filespec = r'\Samples\PanoPositioning\alignment.xml'
# filespec = open(filename)
alignment_poses = read_navvis_alignment_xml(filename)
print(alignment_poses)
```

```python
datasets = [
    'pano-poses - dataset1.csv',
    'pano-poses - dataset2.csv',
    'pano-poses - dataset3.csv',
]

# ppcs = [read_navvis_pano_poses_csv(os.path.join('test_data', dataset)) for dataset in datasets]
ppcs = [read_navvis_pano_poses_csv(os.path.join(dir, dataset)) for dataset in datasets]

for ppc in ppcs:
    ppc.plot()
```

```python
transformed_ppcs = []
for ppc, ap in zip(ppcs, alignment_poses):
    transformed_ppcs.append(ppc.transform(ap))
```

```python
for ppc in transformed_ppcs:
    ppc.plot(headings=True)
```

```python
plot_pose_collections(transformed_ppcs, size=(16, 10))
```
